```python
from Bio import SeqIO
from pathlib import Path
```

```python
# Make sure DATA_DIR exists
DATA_DIR = Path("data/uniprot_sequences")
```

```python
merged_fasta = DATA_DIR / "all_viral_proteins.fasta"
```

```python
with open(merged_fasta, "w") as out_handle:
    for fasta_file in DATA_DIR.glob("*.fasta"):
        if fasta_file.name != "all_viral_proteins.fasta":
            for record in SeqIO.parse(fasta_file, "fasta"):
                SeqIO.write(record, out_handle, "fasta")
```

```python
print(f"[✓] Combined FASTA saved to {merged_fasta}")
```

```python
import subprocess

def run_phmmer(query_fasta, db_fasta, output_tbl):
    cmd = ["phmmer", "--tblout", output_tbl, query_fasta, db_fasta]
    subprocess.run(cmd)
    print(f"phmmer done: {output_tbl}")

# phmmer
run_phmmer("HIV_Nef.fasta", "all_viral_proteins.fasta", "phmmer_Nef.tbl")
run_phmmer("HCV_Nef.fasta", "all_viral_proteins.fasta", "phmmer_Nef.tbl")
run_phmmer("SARS_COV_2_Nef.fasta", "all_viral_proteins.fasta", "phmmer_Nef.tbl")
```

```python
def run_hmmscan(seqs_fasta, hmm_db, output_tbl):
    cmd = ["hmmscan", "--domtblout", output_tbl, hmm_db, seqs_fasta]
    subprocess.run(cmd)
    print(f"hmmscan done: {output_tbl}")

# hmmscan
run_hmmscan("HIV_proteins.fasta", "Pfam-A.hmm", "hmmscan_HIV.tbl")
run_hmmscan("HCV_proteins.fasta", "Pfam-A.hmm", "hmmscan_HCV.tbl")
run_hmmscan("SARS_COV_2_proteins.fasta", "Pfam-A.hmm", "hmmscan_SARS_COV_2.tbl")
```

```python
def run_hmmsearch(hmm_profile, db_fasta, output_tbl):
    cmd = ["hmmsearch", "--tblout", output_tbl, hmm_profile, db_fasta]
    subprocess.run(cmd)
    print(f"hmmsearch done: {output_tbl}")

# hmmsearch
run_hmmsearch("Nef.hmm", "all_viral_proteins.fasta", "hmmsearch_Nef.tbl")
run_hmmsearch("Nef.hmm", "all_viral_proteins.fasta", "hmmsearch_Nef.tbl")
run_hmmsearch("Nef.hmm", "all_viral_proteins.fasta", "hmmsearch_Nef.tbl")
```

```python
def run_jackhmmer(query_fasta, db_fasta, output_tbl, iterations=3):
    cmd = ["jackhmmer", "--tblout", output_tbl, "-N", str(iterations), query_fasta, db_fasta]
    subprocess.run(cmd)
    print(f"jackhmmer done: {output_tbl}")

# Example
run_jackhmmer("HIV_Nef.fasta", "all_viral_proteins.fasta", "jackhmmer_Nef.tbl", iterations=3)
run_jackhmmer("HCV_Nef.fasta", "all_viral_proteins.fasta", "jackhmmer_Nef.tbl", iterations=3)
run_jackhmmer("SARS_COV_2_Nef.fasta", "all_viral_proteins.fasta", "jackhmmer_Nef.tbl", iterations=3)
```
